# **Chance of Admission for Higher Studies**

Predict the chances of admission of a student to a Graduate program based on:

1. GRE Scores (290 to 340)
2. TOEFL Scores (92 to 120)
3. University Rating (1 to 5)
4. Statement of Purpose (1 to 5)
5. Letter of Recommendation Strength (1 to 5)
6. Undergraduate CGPA (6.8 to 9.92)
7. Research Experience (0 or 1)
6. Chance of Admit (0.34 to 0.97)

```python
import pandas as pd
```

```python
admission = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
admission.head()
```

```python
admission.info()
```

```python
admission.describe()
```

```python
# define target (y) and features (X)
```

```python
admission.columns
```

```python
y = admission['Chance of Admit ']
```

```python
X = admission.drop(['Serial No','Chance of Admit '],axis=1)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Create a DataFrame to compare actual vs. predicted values
comparison_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

# Display the first few rows of the comparison
print("Actual vs. Predicted Chance of Admit for the Test Set:")
display(comparison_df.head(10))
```

```python
# model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Actual', y='Predicted', data=comparison_df)
plt.title('Actual vs. Predicted Chance of Admit')
plt.xlabel('Actual Chance of Admit')
plt.ylabel('Predicted Chance of Admit')
# Add a diagonal line for perfect prediction
plt.plot([min(comparison_df['Actual']), max(comparison_df['Actual'])],
         [min(comparison_df['Actual']), max(comparison_df['Actual'])], color='red', linestyle='--')
plt.grid(True)
plt.show()
```

In this scatter plot, each point represents a student from the test set. The red dashed line signifies where predicted values would perfectly match actual values. Points closer to this line indicate better model accuracy. This visualization helps to see the spread and any systematic errors in your predictions.

## Visualize Feature Distributions

**Reasoning**:
To visualize the distributions of the specified features, I will generate histograms for 'GRE Score', 'TOEFL Score', 'CGPA', and 'Chance of Admit' using seaborn's histplot function, ensuring each plot has a descriptive title and appropriate labels.
